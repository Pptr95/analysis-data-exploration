## Speeding up scalar selection

```python
college = pd.read_csv('../section2/data/college.csv', index_col='Unnamed: 0')
```

```python
ac = 'Adrian College'
college.loc[ac, 'Enroll']
```

```python
college.at[ac, 'Enroll']
```

The difference between *loc* and *at* stands in the computation. Let's calculate the time speed using **%timeit** command

```python
%timeit college.loc[ac, 'Enroll']
```

```python
%timeit college.at[ac, 'Enroll']
```

Let's do the same for *iloc* and *iat*

```python
start = college.index.get_loc(ac)
stop = college.columns.get_loc('Enroll')
```

```python
%timeit college.iloc[start, stop]
```

```python
%timeit college.iat[start, stop]
```
